```python
dateColumns=['YearBuilt','YearRemodAdd','GarageYrBlt','YrSold']

for i in dateColumns:
    
    dataset[i]=pd.to_datetime(dataset[i])
        
    print(dataset[i])    
       
```

```python
stringColumns=[]
for i in dataset:
    if (dataset[i].dtype=="str"):
       stringColumns.append(i)        
dataset=pd.get_dummies(dataset,columns=stringColumns,drop_first=True)
print(f"missing values left : {dataset.isnull().sum().sum()}")     
print(f"Non-numeric columns : {dataset.select_dtypes(include='object').columns}")     
```

```python
# 1. تحديد كل الأعمدة اللي نوعها مو رقمي (object أو category)
cols_to_encode = dataset.select_dtypes(exclude=['number']).columns.tolist()

# 2. تحويلهم لـ One-Hot Encoding بضربة وحدة
dataset = pd.get_dummies(dataset, columns=cols_to_encode, drop_first=True)

# 3. فحص أخير: هل لسا في أعمدة نصوص؟
remaining_objects = dataset.select_dtypes(include=['object']).columns.tolist()
print(f"Remaining non-numeric columns: {remaining_objects}")
```
